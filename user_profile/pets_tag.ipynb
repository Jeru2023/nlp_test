```python
import warnings
from user_profile.profile_service import *
warnings.filterwarnings('ignore')

INPUT_FILE = 'data/xhs-T1-40W.csv'
CONFIG_FILE = 'config/config.yaml'
DOMAIN = 'PETS'
```

```python
# load config
config = load_config(CONFIG_FILE)

global_nickname_blacklist = get_global_nickname_blacklist(config)
domain_nickname_blacklist = get_domain_nicknames_blacklist(config, domain=DOMAIN)

domain_content_whitelist = get_content_whitelist(config, domain=DOMAIN)
domain_content_blacklist = get_content_blacklist(config, domain=DOMAIN)

cat_content_whitelist = get_content_whitelist(config, domain=DOMAIN, sub_domain='cat')
cat_content_blacklist = get_content_blacklist(config, domain=DOMAIN, sub_domain='cat')

dog_content_whitelist = get_content_whitelist(config, domain=DOMAIN, sub_domain='dog')
dog_content_blacklist = get_content_blacklist(config, domain=DOMAIN, sub_domain='dog')
```

```python
# load data then remove global biz user
df = load_data(INPUT_FILE)
df = remove_biz_user(df, global_nickname_blacklist)
len(df)
```

```python
# filter by score based on content white/black list and score threshold
df_filtered = filter_by_score(df, domain_content_whitelist, domain_content_blacklist, 5)
df_user = group_by_user(df_filtered)
print(f'pets - #posts:{len(df_filtered)}, #users: {len(df_user)}')
```

```python
# find cat users
df_filter_cat = filter_by_score(df_filtered, cat_content_whitelist, cat_content_blacklist, 5)
df_user_cat = group_by_user(df_filter_cat)
print(f'cat - #posts:{len(df_filter_cat)}, #users: {len(df_user_cat)}')
```

```python
# find dog users
df_filter_dog = filter_by_score(df_filtered, dog_content_whitelist, dog_content_blacklist, 5)
df_user_dog = group_by_user(df_filter_dog)
print(f'dog - #posts:{len(df_filter_dog)}, #users: {len(df_user_dog)}')
```
