```python
import warnings
from user_profile.profile_service import *
warnings.filterwarnings('ignore')

INPUT_FILE = 'data/xhs-T1-40W.csv'
CONFIG_FILE = 'config/config.yaml'
DOMAIN = 'BABY'
```

```python
# load config
config = load_config(CONFIG_FILE)

global_nickname_blacklist = get_global_nickname_blacklist(config)
domain_nickname_blacklist = get_domain_nicknames_blacklist(config, domain=DOMAIN)

domain_content_whitelist = get_content_whitelist(config, domain=DOMAIN)
domain_content_blacklist = get_content_blacklist(config, domain=DOMAIN)

infant_content_whitelist = get_content_whitelist(config, domain=DOMAIN, sub_domain='infant')
infant_content_blacklist = get_content_blacklist(config, domain=DOMAIN, sub_domain='infant')

child_content_whitelist = get_content_whitelist(config, domain=DOMAIN, sub_domain='child')
child_content_blacklist = get_content_blacklist(config, domain=DOMAIN, sub_domain='child')

juvenile_content_whitelist = get_content_whitelist(config, domain=DOMAIN, sub_domain='juvenile')
juvenile_content_blacklist = get_content_blacklist(config, domain=DOMAIN, sub_domain='juvenile')
```

```python
# load data then remove global biz user
df = load_data(INPUT_FILE)
df = remove_biz_user(df, global_nickname_blacklist)
df = remove_biz_user(df, domain_nickname_blacklist)
len(df)
```

```python
# filter by score based on content white/black list and score threshold
df_filtered = filter_by_score(df, domain_content_whitelist, domain_content_blacklist, 5)
df_user = group_by_user(df_filtered)
print(f'Child - #posts:{len(df_filtered)}, #users: {len(df_user)}')
```
